### 02. If statement with user-defined function

```python
#Define price_label function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Create new 'price_range' column for function output - apply function on every row (axis = 1)
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Check value_counts for new column
df['price_range'].value_counts()
```

```python
#Check for highest price
df['prices'].max()
```

### 03. Create 'busiest_day' column using if-statement with for-loop

```python
#Check orders_day_of_the_week value counts
df_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Use for-loop to create 'busiest_day' values list - based on orders_day_of_week values
result = []

for value in df_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#'result' is a list that can be inserted into new column
result
```

```python
#Create new column 'busiest_day' with list
df_merged['busiest_day'] = result
```

```python
#Check value counts for new column
df_merged['busiest_day'].value_counts(dropna = False)
```

These counts line up with orders_day_of_week counts.

### 04. Create 'two_busiest_days' column using if-statement with for-loop

```python
#Check orders_day_of_week counts for second busiest and second least busy days of week
df_merged['orders_day_of_week'].value_counts(dropna = False)
```

0 and 1 are busiest days, 3 and 4 are least busy days.

```python
#Use for-loop to create 'two_busiest_days' values list - based on orders_day_of_week values
result = []

for value in df_merged["orders_day_of_week"]:
  if value in [0,1]:
    result.append("Busiest days")
  elif value in [3,4]:
    result.append("Least busy days")
  else:
    result.append("Regularly busy")
```

```python
#Create new column for 'two_busiest_days'
df_merged['two_busiest_days'] = result
```

```python
#Check value counts
df_merged['two_busiest_days'].value_counts()
```

These value counts line up with orders_day_of_week value counts

### 05. Create 'busiest_period_of_day' column using if-statement with for-loop

```python
#Check 'order_hour_of_day' value counts
df_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#Use for-loop to create 'busiest_period_of_day' values list - based on order_hour_of_day values
result = []

for value in df_merged["order_hour_of_day"]:
  if value in range(9,17):
    result.append("Most orders")
  elif value in [17,8,18,19,20,7,21,22]:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

```python
#Assign result list to 'busiest_period_of_day' column
df_merged['busiest_period_of_day'] = result
```

```python
#Check value counts
df_merged['busiest_period_of_day'].value_counts(dropna = False)
```

These value counts line up with with order_hour_of_day value counts. 'Most orders' sum equals sum of top 8 hours of day.
